# **Designing a toy stream cipher using LFSR**

### Question 10: Build your own stream cipher combining two LFSR with primitive polynomials. The output of the two LFSRs are combined with a Xor operation. Explain you code.
* On the sender side (ALICE Side) encrypt a message using the generated keystream
* Then, on the receiver side (BOB Side) decrypt the message and check if the original message is retrieved

````python
from pylfsr import LFSR

# LFSR 1 (3-bit, primitive polynomial x^3 + x + 1)
LFSR_1 = LFSR(initstate=[1,1,1], fpoly=[3,1])

# LFSR 2 (4-bit, primitive polynomial x^4 + x + 1)
LFSR_2 = LFSR(initstate=[1,0,1,1], fpoly=[4,1])

# Generate 15-bit keystream by XORing outputs
keystream = []
for _ in range(15):
    bit = LFSR_1.next() ^ LFSR_2.next()
    keystream.append(int(bit))  

print("Keystream:", keystream)

message = "Bonjour"

# Encryption
cipher = ''.join(chr(ord(m) ^ k) for m, k in zip(message, keystream))
print("Ciphertext:", cipher)

# Decryption
decrypted = ''.join(chr(ord(c) ^ k) for c, k in zip(cipher, keystream))
print("Decrypted message:", decrypted)
````

### Question 11: Using the previous stream cipher, on the sender side, write a program that reads a file and encrypts the content of the file to provide a ciphertext that is stored in a new file. On the receiver side, read the ciphertext from the new file and decrypt the content of the file to retrieve the plaintext.


````python
from pylfsr import LFSR
from pathlib import Path

# ----------------------------
# LFSR CONFIGURATION (must be same on sender & receiver)
# ----------------------------
LFSR_1_INIT = [1, 1, 1]
LFSR_1_FPOLY = [3, 1]         # x^3 + x + 1

LFSR_2_INIT = [1, 0, 1, 1]
LFSR_2_FPOLY = [4, 1]         # x^4 + x + 1

def create_lfsrs():
    """Create and return two LFSR objects."""
    LFSR_1 = LFSR(initstate=LFSR_1_INIT, fpoly=LFSR_1_FPOLY)
    LFSR_2 = LFSR(initstate=LFSR_2_INIT, fpoly=LFSR_2_FPOLY)
    return LFSR_1, LFSR_2

def generate_keystream(num_bytes):
    """Generate keystream of N bytes by XORing outputs of both LFSRs."""
    LFSR_1, LFSR_2 = create_lfsrs()
    keystream = bytearray()

    for _ in range(num_bytes):
        byte = 0
        for _ in range(8):  # 8 bits per byte
            bit = int(LFSR_1.next()) ^ int(LFSR_2.next())
            byte = (byte << 1) | bit
        keystream.append(byte)

    return bytes(keystream)

def xor_bytes(data, keystream):
    """XOR two byte sequences and return the result."""
    return bytes(d ^ k for d, k in zip(data, keystream))

def encrypt_file(input_file, output_file):
    """ALICE: Read plaintext file → XOR with keystream → save ciphertext."""
    plaintext = Path(input_file).read_bytes()
    ks = generate_keystream(len(plaintext))
    ciphertext = xor_bytes(plaintext, ks)
    Path(output_file).write_bytes(ciphertext)
    print(f"[ALICE] Encrypted '{input_file}' → '{output_file}'")

def decrypt_file(input_file, output_file):
    """BOB: Read ciphertext file → XOR with same keystream → recover plaintext."""
    ciphertext = Path(input_file).read_bytes()
    ks = generate_keystream(len(ciphertext))
    plaintext = xor_bytes(ciphertext, ks)
    Path(output_file).write_bytes(plaintext)
    print(f"[BOB] Decrypted '{input_file}' → '{output_file}'")

if __name__ == "__main__":
    # Example file names
    plaintext_file = "message.txt"
    ciphertext_file = "encrypted.bin"
    decrypted_file = "decrypted.txt"

    # Create test file if not present
    if not Path(plaintext_file).exists():
        Path(plaintext_file).write_text("HELLO THIS IS A TEST OF LFSR STREAM CIPHER")

    # Encrypt and decrypt
    encrypt_file(plaintext_file, ciphertext_file)
    decrypt_file(ciphertext_file, decrypted_file)

    # Check integrity
    if Path(plaintext_file).read_bytes() == Path(decrypted_file).read_bytes():
        print("✅ Decryption successful — original message recovered.")
    else:
        print("❌ Error — decrypted file does not match original.")
````
